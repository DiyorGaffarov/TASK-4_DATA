```python
string_columns = orders_df.select_dtypes(include='object').columns

pd.DataFrame({
    'empty_string': (orders_df[string_columns] == '').sum(),
    'NaN': orders_df[string_columns].isnull().sum(),
    'total_missing': ((orders_df[string_columns] == '').sum() + orders_df[string_columns].isnull().sum())
})
```
