```python
import pandas as pd 
import yaml

users_df = pd.read_csv("../data3/users.csv") 

with open("../data3/books.yaml") as books:
    data=yaml.safe_load(books) 
books_df = pd.DataFrame(data)

orders_df = pd.read_parquet("../data3/orders.parquet")
```

```python
print("Users df: " + str(users_df.duplicated().sum()))
print("Books df: " + str(books_df.duplicated().sum()))
print("Orders df: " + str(orders_df.duplicated().sum()))
```

```python
print("Users df: "+str(users_df.duplicated(subset=[col for col in users_df.columns if col != 'id']).sum()))
print("Books df: "+str(books_df.duplicated(subset=[col for col in books_df.columns if col != 'id']).sum()))
print("Orders df: "+str(orders_df.duplicated(subset=[col for col in orders_df.columns if col != 'id']).sum()))
```
